```
import numpy as np
import time

a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a,b)
toc = time.time()
print(c)

print("Vectorized version "+str(1000*(toc-tic))+"ms")
```

```
c = 0
tic = time.time()
for i in range(len(a)):
  c+=a[i]*b[i]
toc = time.time()
print(c)

print("Non-Vectorized version "+str(1000*(toc-tic))+"ms")
```
